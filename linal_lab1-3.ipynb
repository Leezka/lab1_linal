```python
def palindromic_squares_and_circular_primes() -> tuple[List[int], List[int]]:
  """
  Возвращает:
      tuple:
      - список всех палиндромов a < 100000, для которых a^2 — палиндром;
      - список всех простых p < 1000000, все циклические перестановки цифр которых
      → просты.
  """
  def is_prime(n: int) -> bool:
    "Проверяет, является ли число простым"
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

  def is_palindromic(n: int) -> bool:
    "Проверяет, является ли число палиндромом"
    s = str(n)
    return s == s[::-1]

  def cyclic_permutations(s: str):
    "Перебирает все циклические перестановки"
    for i in range(len(s)):
            yield s[i:] + s[:i]

  "Решение первого пункта"
  palindromic_squares = [
  a for a in range(1, 10**5)
  if is_palindromic(a) and is_palindromic(a**2)
  ]

  "Решение второго пункта"
  circular_primes = []
  for p in range(2, 10**6):
    if not is_prime(p):
      continue
    s = str(p)
    if all(is_prime(int(rot)) for rot in cyclic_permutations(s)):
       circular_primes.append(p)

  return (palindromic_squares, circular_primes)
print(palindromic_squares_and_circular_primes())
```

```python
def primes_with_two_digits() -> Dict[str, List[int]]:
    """
    Возвращает словарь вида:
    {
      '13': [список первых 100 простых из {1,3}],
      '15': [список первых 100 простых из {1,5}],
      '17': [список первых 100 простых из {1,7}],
      '19': [список первых 100 простых из {1,9}],
    }
    """
    def is_prime(n: int) -> bool:
      "Проверяет, является ли число простым"
      if n < 2:
        return False
      for i in range(2, int(n**0.5) + 1):
         if n % i == 0:
            return False
      return True
    digit_pairs = ['13', '15', '17', '19']
    result = {pair: [] for pair in digit_pairs}
    for pair in digit_pairs:
        digits = list(pair)
        length = 2
        while len(result[pair]) < 100:
            for p in product(digits, repeat=length):
                num = int(''.join(p))
                if is_prime(num):
                    result[pair].append(num)
                    if len(result[pair]) == 100:
                        break
            length += 1
    return result

print(primes_with_two_digits())


```

```python
def factorial_plus_one_factors() -> Dict[int, Dict[int, int]]:
    """
    Возвращает словарь вида:
        { n: (простой _делитель: степень, ...], ... }
    для п от 2 до 50, где ключ - п, значение - разложение п + 1 на простые множители.
    """
    result = {}
    factorial = 1
    for n in range(1, 51):
        factorial *= n
        if n >= 2:
            result[n] = factorint(factorial + 1)
    max_primes = max(len(factors) for factors in result.values())
    big_primes = [(n, max(factors)) for n, factors in result.items() if max(factors) > 10**6]
    return max_primes, big_primes

max_primes, big_primes = factorial_plus_one_factors()
print("Ответ на первый пункт:", max_primes)
print("Ответ на второй пункт:", big_primes)
```

```python
from typing import List, Dict
from math import gcd
import time
from sympy import factorint, totient

def euler_phi_direct(n: int) -> int:
    """
    Функция Эйлера через полный перебор чисел, взаимно простых с n.
    """
    counter = 0
    for candidate in range(1, n + 1):
        if gcd(candidate, n) == 1:
            counter += 1
    return counter

def euler_phi_factor(n: int) -> int:
    """
    Функция Эйлера через разложение на простые множители.
    """
    prime_factors = factorint(n)  # словарь {простое число: степень}
    phi = n
    for prime in prime_factors:
        phi *= (1 - 1 / prime)
    return int(phi)

def compare_euler_phi_methods(test_values: List[int]) -> Dict[str, List[float]]:
    """
    Сравнение трёх методов вычисления функции Эйлера по времени выполнения.
    """
    times_direct: List[float] = []
    times_factor: List[float] = []
    times_sympy: List[float] = []

    for number in test_values:
        # прямой метод
        start_time = time.time()
        euler_phi_direct(number)
        times_direct.append(time.time() - start_time)

        # через простые множители
        start_time = time.time()
        euler_phi_factor(number)
        times_factor.append(time.time() - start_time)

        # встроенная функция SymPy
        start_time = time.time()
        totient(number)
        times_sympy.append(time.time() - start_time)

    return {
        "direct": times_direct,
        "factor": times_factor,
        "sympy": times_sympy
    }

if __name__ == "__main__":
    numbers_to_test = [10, 100, 1000, 5000, 10000]

    results = compare_euler_phi_methods(numbers_to_test)

    print("Тестовые числа:", numbers_to_test)
    print("Время выполнения методов Эйлера (секунды):")
    print("Прямой перебор:", results["direct"])
    print("Через разложение на простые множители:", results["factor"])
    print("SymPy totient:", results["sympy"])
```

```python
from itertools import permutations
from typing import List, Tuple, Set
```

```python
def permute(p: Tuple[int, ...], q: Tuple[int, ...]) -> Tuple[int, ...]:
    return tuple(p[q[i]-1] for i in range(len(p)))

def build_subgroup(gens: List[Tuple[int, ...]], group: List[Tuple[int, ...]]) -> Set[Tuple[int, ...]]:
    subgroup = {tuple(range(1, len(group[0]) + 1))}
    changed = True
    while changed:
        changed = False
        for g in gens:
            for h in list(subgroup):
                new_elem = permute(g, h)
                if new_elem not in subgroup:
                    subgroup.add(new_elem)
                    changed = True
    return subgroup

def left_cosets(H: Set[Tuple[int, ...]], G: List[Tuple[int, ...]]) -> List[Set[Tuple[int, ...]]]:
    """Строит левый смежный класс подгруппы H в G."""
    remaining = set(G)
    cosets = []
    while remaining:
        g = remaining.pop()
        coset = {permute(g, h) for h in H}
        cosets.append(coset)
        remaining -= coset
    return cosets

def right_cosets(H: Set[Tuple[int, ...]], G: List[Tuple[int, ...]]) -> List[Set[Tuple[int, ...]]]:
    """Строит правый смежный класс подгруппы H в G."""
    remaining = set(G)
    cosets = []
    while remaining:
        g = remaining.pop()
        coset = {permute(h, g) for h in H}
        cosets.append(coset)
        remaining -= coset
    return cosets

def subgroups_of_Sm(N: int) -> dict:
    """
    Находит все подгруппы группы S_m и выбирает одну для анализа.
    """
    m = 4
    G = [tuple(p) for p in permutations(range(1, m + 1))]
    subgroup_set = set()
    for r in [1, 2, 3]:
        for gens in permutations(G, r):
            H = build_subgroup(list(gens), G)
            subgroup_set.add(frozenset(H))

    subgroups = [set(s) for s in subgroup_set]
    chosen_H = subgroups[N % len(subgroups)]
    L = left_cosets(chosen_H, G)
    R = right_cosets(chosen_H, G)

    return {
        "m": m,
        "total_subgroups": len(subgroups),
        "chosen_subgroup": chosen_H,
        "left_cosets": L,
        "right_cosets": R,
        "index": len(G) // len(chosen_H),
        "is_normal": L == R
    }

N = 15
result = subgroups_of_Sm(N)

print(f"Количество подгрупп: {result['total_subgroups']}")
print(f"Выбранная подгруппа: {result['chosen_subgroup']}")
print(f"Левые смежные классы: {result['left_cosets']}")
print(f"Правые смежные классы: {result['right_cosets']}")
print(f"Индекс подгруппы: {result['index']}")
print(f"Является нормальной: {result['is_normal']}")
```

```python
def element_powers_in_Sm(N: int) -> dict:
    m_val = 4  # порядок симметрической группы
    exp1, exp2, exp3 = 3, 4, 5  # степени
    sym_group = SymmetricGroup(m_val)
    group_elems = list(sym_group.elements)  # все элементы группы

    idx = N % len(group_elems)
    base_elem = group_elems[idx]  # выбранный элемент по индексу

    # возводим элемент в степени
    pow1, pow2, pow3 = base_elem ** exp1, base_elem ** exp2, base_elem ** exp3

    # порядок каждого возведённого элемента
    ord1, ord2, ord3 = pow1.order(), pow2.order(), pow3.order()

    # подгруппы, порожденные этими элементами
    sub1, sub2, sub3 = sym_group.subgroup([pow1]), sym_group.subgroup([pow2]), sym_group.subgroup([pow3])
    sub_ord1, sub_ord2, sub_ord3 = sub1.order(), sub2.order(), sub3.order()

    return {
        'base_element': {
            'repr': str(base_elem),
            'order_val': base_elem.order()
        },
        'power1': {
            'exp': exp1,
            'elem_order': ord1,
            'sub_order': sub_ord1
        },
        'power2': {
            'exp': exp2,
            'elem_order': ord2,
            'sub_order': sub_ord2
        },
        'power3': {
            'exp': exp3,
            'elem_order': ord3,
            'sub_order': sub_ord3
        }
    }

res = element_powers_in_Sm(15)
print(f"Базовый элемент: {res['base_element']['repr']}, порядок: {res['base_element']['order_val']}")
print(f"g^{res['power1']['exp']}: порядок элемента {res['power1']['elem_order']}, порядок подгруппы {res['power1']['sub_order']}")
print(f"g^{res['power2']['exp']}: порядок элемента {res['power2']['elem_order']}, порядок подгруппы {res['power2']['sub_order']}")
print(f"g^{res['power3']['exp']}: порядок элемента {res['power3']['elem_order']}, порядок подгруппы {res['power3']['sub_order']}")

```

```python
def solve_sigma_power_eq(N: int) -> dict:
    m, n = 4, 7  # порядок группы и степень
    G = SymmetricGroup(m)
    target = Permutation(list(range(m - 1)))  # целевой цикл

    # находим все элементы σ, такие что σ^n = target
    sols = [s for s in G.generate_schreier_sims() if s**n == target]

    # выбираем три случайных решения
    rand_sols = random.sample(sols, min(3, len(sols)))
    if sols:
        orders = [s.order() for s in sols]  # порядки решений
        cycles = [s.cycle_structure for s in sols]  # структуры циклов
        order_info = f"Все решения имеют порядок {orders[0]}" if len(set(orders))==1 else f"Порядки решений: {sorted(set(orders))}"
        cycle_info = "Все решения имеют одинаковую цикловую структуру" if len(set(map(str, cycles)))==1 else "Решения имеют различную цикловую структуру"
    else:
        order_info, cycle_info = "Решений не найдено", ""

    return {
        'equation': f"σ^{n} = {target}",
        'total_solutions': len(sols),
        'random_solutions': [str(s) for s in rand_sols],
        'properties': [order_info, cycle_info]
    }

res = solve_sigma_power_eq(15)
print(f"Уравнение: {res['equation']}")
print(f"Количество решений: {res['total_solutions']}")
print(f"Примеры решений: {res['random_solutions']}")
for p in res['properties']:
    if p: print(f"- {p}")
```

```python
from math import gcd

def isomorphism_of_cyclic_subgroup_Zm_star(N: int) -> dict:
    m, t = 4, 9
    # Элементы мультипликативной группы Z_m*
    Zm_star = [i for i in range(1, m) if gcd(i, m) == 1]
    # Циклическая подгруппа, порожденная t mod m
    t_mod_m = t % m
    subgroup = set()
    current = 1
    while current not in subgroup:
        subgroup.add(current)
        current = (current * t_mod_m) % m
        if current == 0:
            break
    subgroup = sorted(subgroup)
    order = len(subgroup)
    # Изоморфная циклическая подгруппа в S_d
    isomorphic_to = f"C_{order}"
    return {
        'm': m,
        't': t,
        't_mod_m': t_mod_m,
        'Zm_star': Zm_star,
        'subgroup': subgroup,
        'order': order,
        'isomorphic_to': isomorphic_to
    }

res = isomorphism_of_cyclic_subgroup_Zm_star(15)
print(f"Циклическая подгруппа: {res['subgroup']}")
print(f"Порядок подгруппы: {res['order']}")
print(f"Изоморфна циклической подгруппе: {res['isomorphic_to']}")
```

```python
x = symbols('x')
def factor_polynomials(num):
    n_val = num % 20

    # коэффициенты полиномов
    c_vals = [(i + n_val) % 5 for i in range(6)]
    d_vals = [(l + n_val) % 9 for l in range(5)]

    # полином 1 в F5
    coeffs1 = [1] + c_vals[:5]
    poly1 = Poly(coeffs1, x, domain=GF(5))
    poly1_expr = poly1.as_expr()  # превращаем в обычное выражение
    factors1 = factor(poly1, modulus=5)
    status1 = "неприводим" if len(factors1.args) == 2 else "приводим"

    # полином 2 в F3
    coeffs2 = [1] + [d % 3 for d in d_vals]
    poly2 = Poly(coeffs2, x, domain=GF(3))
    poly2_expr = poly2.as_expr()
    factors2 = factor(poly2, modulus=3)
    status2 = "неприводим" if len(factors2.args) == 2 else "приводим"
    return {
        'poly1': poly1_expr,
        'status1': status1,
        'factors1': factors1.as_expr(),
        'poly2': poly2_expr,
        'status2': status2,
        'factors2': factors2.as_expr()
    }

res = factor_polynomials(504655)
print("Полином 1 (F₅):", res['poly1'])
print("Статус:", res['status1'])
print("Разложение:", res['factors1'])
print()
print("Полином 2 (F₃):", res['poly2'])
print("Статус:", res['status2'])
print("Разложение:", res['factors2'])
```

```python
from sympy import symbols, Poly, GF

x = symbols('x')

def solve_gcd_problem():
    # коэффициенты полиномов из условия
    rm = [4, 5, 6, 7, 8, 9]
    st = [4, 5, 6, 7, 8, 9]

    # формируем полиномы f(x) и g(x)
    f_coeffs = rm + [rm[0], rm[1]]  # f(x) степени 7
    g_coeffs = st[:4]                # g(x) степени 3

    # создаём полиномы над F11
    F11 = GF(11)
    f_poly = Poly(f_coeffs, x, domain=F11)
    g_poly = Poly(g_coeffs, x, domain=F11)

    # gcd и линейное представление: u*f + v*g = gcd
    u_poly, v_poly, gcd_poly = f_poly.gcdex(g_poly)

    # функция для вывода полинома
    def format_poly(poly):
        coeffs = [int(c) % 11 for c in poly.all_coeffs()]
        terms = []
        deg = len(coeffs) - 1
        for i, c in enumerate(coeffs):
            if c == 0:
                continue
            power = deg - i
            if power == 0:
                terms.append(f"{c}")
            elif power == 1:
                terms.append(f"{'' if c==1 else c}x")
            else:
                terms.append(f"{'' if c==1 else c}x^{power}")
        return " + ".join(terms) if terms else "0"

    print("gcd(f,g) =", format_poly(gcd_poly))
    print("u(x) =", format_poly(u_poly))
    print("v(x) =", format_poly(v_poly))

solve_gcd_problem()
```

```python
from sympy import symbols, Poly, GF, invert
from sympy.abc import x

def solve_inverse_problem():
    # Параметры из условия (s_i из st с N=15)
    st = [4, 5, 6, 7, 8, 9]
    s0, s1, s2 = st[0] % 13, st[1] % 13, st[2] % 13

    # Полиномы над F13
    F13 = GF(13)
    f_poly = Poly([s2, s1, s0], x, domain=F13)  # s2*x^2 + s1*x + s0
    g_poly = Poly([1, 0, 0, 0, 1, 1, 0, 6, 2], x, domain=F13)  # x^8 + x^4 + x^3 + 6x + 2

    # Находим обратный элемент
    h_poly = invert(f_poly, g_poly)

    # Форматируем ответ
    def format_poly(coeffs):
        terms = []
        degree = len(coeffs) - 1
        for i, coeff in enumerate(coeffs):
            power = degree - i
            coeff_int = int(coeff) % 13
            if coeff_int == 0:
                continue
            if power == 0:
                terms.append(f"{coeff_int}")
            elif power == 1:
                terms.append(f"{coeff_int}x" if coeff_int != 1 else "x")
            else:
                terms.append(f"{coeff_int}x^{power}" if coeff_int != 1 else f"x^{power}")
        return " + ".join(terms) if terms else "0"

    print(f"h(x) = {format_poly(h_poly.all_coeffs())}")

solve_inverse_problem()
```

```python
from sympy import symbols, Poly, GF
from sympy.abc import x
from itertools import product

def generate_irreducible_polynomials(q: int, d: int) -> list:
    Fq = GF(q)
    irreducible = []

    for coeffs in product(range(q), repeat=d+1):
        if coeffs[0] == 0:  # старший коэффициент не может быть 0
            continue
        poly = Poly(coeffs, x, domain=Fq)
        if poly.is_irreducible:
            irreducible.append(poly)

    return irreducible

def format_poly(poly):
    coeffs = poly.all_coeffs()
    terms = []
    degree = len(coeffs) - 1
    for i, coeff in enumerate(coeffs):
        power = degree - i
        coeff_int = int(coeff)
        if coeff_int == 0:
            continue
        if power == 0:
            terms.append(f"{coeff_int}")
        elif power == 1:
            terms.append(f"x" if coeff_int == 1 else f"{coeff_int}x")
        else:
            terms.append(f"x^{power}" if coeff_int == 1 else f"{coeff_int}x^{power}")
    return " + ".join(terms) if terms else "0"

def test_irreducible_polynomials():
    test_cases = [
        (2, 2), (2, 3), (2, 4),
        (3, 2), (3, 3), (3, 4),
        (5, 2), (5, 3), (5, 4)
    ]
    for q, d in test_cases:
        print(f"F_{q}, степень {d}:")
        polys = generate_irreducible_polynomials(q, d)
        for poly in polys:
            print(f"  {format_poly(poly)}")
        print(f"  Всего: {len(polys)}\n")

if __name__ == "__main__":
    test_irreducible_polynomials()
```
